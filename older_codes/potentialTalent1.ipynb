```python
### using gensim for word embeddings and pandas for data manipulation
import gensim
import pandas as pd
```

```python
### loading the data
df_pt = pd.read_csv("../data/potential-talents.csv")

```

```python
job_title=df_pt['job_title'].apply(gensim.utils.simple_preprocess)
```

```python
job_title.head()
```

```python
job_title.loc[0]
```

```python
model = gensim.models.Word2Vec(sentences=job_title, vector_size=100, window=10, min_count=2, workers=4)
```

```python
model.build_vocab(job_title, progress_per=1000)
```

```python
model.train(df_pt['job_title'], total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.save("word2vec_potentialTalent.model")
```
