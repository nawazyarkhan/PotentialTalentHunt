## notebook created for testing fastapi endpoints

```python
import requests
import json
import pandas as pd
```

```python
url = 'http://127.0.0.1:8000/health'
```

```python
response = requests.get(url)
print(response.json())
print(response.status_code)
```

```python
url = 'http://127.0.0.1:8000/search_jds'
query = "HR manager" 
```

```python
params = {    "query": query,
    "top_k": 10,
    "threshold": 0.7
}
%time response = requests.post(url, data=json.dumps(params))

# Check if server is responding
print(f"Status Code: {response.status_code}")
print(f"Response Text: {response.text}")

# Only try to parse JSON if response is successful
if response.status_code == 200:
    print(response.json())
else:
    print(f"Error: Server returned status {response.status_code}")
    print("Make sure the FastAPI server is running at http://127.0.0.1:8000")
```

```python
# converting response in json format to pretty print
response_json = response.json()
print(json.dumps(response_json, indent=4))
```

```python
## converting json response to pandas dataframe

df = pd.json_normalize(response_json['results'])
df.head(10)
```
